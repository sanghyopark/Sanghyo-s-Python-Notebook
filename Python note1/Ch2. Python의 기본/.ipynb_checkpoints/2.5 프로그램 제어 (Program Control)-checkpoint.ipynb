```python
points = [86, 74, 95, 65, -35, 81]
for n, point in enumerate(points):
    if point < 0:
        print('Something wrong in the data!')
        break
    elif point < 80:
        print(f'The {n}th student with point={point} has failed.')
        continue
    else:
        pass
    print(f'The {n}th student with point={point} has passed.')
```

```python
La = [0, -1, 5, -5, 3, 7]

La1 = []
for x in La:
    if 0<=x<=5: La1.append(x)
print(La1)

La2 = La.copy()
for x in La:
    if x<0 or x>5: La2.remove(x)
print(La2)

La3 = [x for x in La if 0<=x<=5]; print(La3)

La4 = list(filter(lambda x: 0<=x<=5, La)); print(La4)

La5 = [5 if x>5 else x for x in [-4 if x<-4 else x for x in La]]; print(La5)

La6 = []
for x in La:
    if x<-4:
        x=-4
        La6.append(x)
    elif x>5:
        x=5
        La6.append(x)
    else:
        La6.append(x)
        
print(La6)
    
```

##  예제 2.4 Poisson 분포 확률의 합

```python
import math
lam = 2
term = math.exp(-lam)
S = 0
for k in range(100):
    S = S + term
    term = term*lam/(k+1)
print(S)
```

## Remark 2.11 for loop의 반복범위에 대해 zip()이나 enumerate()함수를 사용하기

```python
fruits = ["Apple", "Banana", "Cherry"]
i = 0
for 과일 in fruits:
    i += 1
    print(f'과일모음[{i:2d}] = {과일}')
```

```python
for i, 과일 in zip(range(1,len(fruits)+1), fruits):
    print(f'과일모음[{i:2d}] = {과일}')
```

```python
for i, 과일 in enumerate(fruits):
    print(f'과일모음[{i:2d}] = {과일}')
```

## 예제 2.5 Loop와 else문의 조화

```python
primes = []
for n in range(2, 10):
    flag = 0
    for x in range(2, n):
        if n%x == 0:
            flag = 1
            break
    if flag ==0:
        primes.append(n)
print('prime numbers: ', primes)
```

```python
primes = []
for n in range(2, 10):
    for x in range(2, n):
        if n%x == 0:
            break
    else:
        primes.append(n)
print('prime numbers: ',primes)
```

# 2.5.3 while Loop

```python
points = [86, 74, 65, 81]
no = 0
while points: #list에 있는 값이 하나씩 빠져나가 False가 나올때까지
    point = points[0]
    no = no+1
    if point >= 80:
        result = 'pass'
    else:
        result = 'fail'
    print(f'{no:2d} 번째 학생: {result}.')
    del points[0]
```

```python
points = [86, 74, 65, 81]
no = 0
while no < len(points):
    point = points[no]
    no = no+1
    if point >= 80:
        result = 'pass'
    else:
        result = 'fail'
    print(f'{no:2d} 번째 학생: {result}.')
```

```python
# 최소 양수를 얻기 위해 2로 나누는 것을 반복한다.
x = 1
while x/2 > 0:
    x = x/2
x_min = x
print(f'x_min ={x_min: 27.20e}')

# 최대 양수를 얻기 위해 2를 곱하는 것을 반복한다.
x = 1
inf = float('inf')
while 2*x < inf:
    x = x*2
x_max0 = x
print(f'x_max0 ={x_max0: 27.20e}')

tmp = 1
while x_max0*(2-tmp/2) < inf:
    tmp = tmp/2
x_max = x*(2-temp)
print(f'x_max ={x_max: 27.20e}')

# %e: 지수로 표현
```

```python
import math
while True:
    r = input('\n 구의 반경: ')
    if r == '': #사용자가 바로 enter를 누른 경우
        break
    if not r.isnumeric(): # r이 수로 변환 가능한 문자인가?
        print('제발 수를, 그것도 0 이상을 주세요!')
        continue
    r = float(r)
    v = 4/3*math.pi*r**3
    print(f'반경이 {r:.2f}인 구의 부피는 {v:.2f}입니다.')
```
