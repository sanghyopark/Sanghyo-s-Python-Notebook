# 8.10 Error Bar 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(tight_layout=True, figsize=(10,3))
x = np.arange(0.1,4,0.5)
y = 1-np.exp(-x)
xe = 0.1 + 0.5*np.log10(1+x) # x에 대한 오차 한계
ye = 0.1*np.sqrt(y) # y에 대한 오차 한계

plt.subplot(1,3,1)
plt.errorbar(x,y, yerr=ye)
plt.title('Vertical errorbar plot', fontsize=10)

plt.subplot(1,3,2)
plt.errorbar(x,y, xerr=xe, errorevery=(1,2))
#errorevery=(1,2): sample중 1번째 sample부터 2개의 sample마다 errorbar을 삽입
plt.title('Horizontal errorbar plot', fontsize=10)

plt.subplot(1,3,3)
plt.errorbar(x,y, yerr=ye, xerr=xe, errorevery=(2,4), fmt='o--')
#errorevery=(1,2): sample중 2번째 sample부터 4개의 sample마다 errorbar을 삽입
#fmt='o--': errorbar type
plt.title('Vertical/Horizontal errorbar plot', fontsize=10)
plt.grid()
plt.show()
```
