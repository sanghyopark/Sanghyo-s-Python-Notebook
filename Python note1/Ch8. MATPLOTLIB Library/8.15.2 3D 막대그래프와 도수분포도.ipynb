# 8.15.2 3D 막대그래프와 도수분포도

```python
def bar3d_my(X, Y, Z, wx=None, wy=None, ax=None):
    import matplotlib.pyplot as plt
    import numpy as np
    
    X = np.array(X); Y = np.array(Y)
    dx = X[0][1]-X[0][0]; dy = Y[1][0]-Y[0][0]
    
    #막대들의 x,y-축 방향 두께
    if wx == None: wx = dx/2
    if wy == None: wy = dy/2
    wx = min([wx, dx]); wy = min([wy, dy])
    xoffset = wx/2; yoffset = wy/2
    Xpos, Ypos = X-xoffset, Y-yoffset
    xpos = Xpos.ravel(); ypos = Ypos.ravel(); zpos = 0
    Z = np.array(Z)
    wz = Z.ravel()
    
    if ax==None:
        ax = plt.subplot(111, projection='3d')
    ax.bar3d(xpos,ypos,0,wx,wy,wz) #, zsort='avarage')
    plt.show()
```

```python
def hist3d_my(x, y, Nb=10, range=None, ax=None):
    # Nb: Number of bins
    import matplotlib.pyplot as plt
    import numpy as np
    
    if range==None: range=[[min(x),max(x)],[min(y),max(y)]]
    zs, xs, ys, = np.histogram2d(x,y, Nb, range)
    dx = xs[1]-xs[0]; dy = ys[1]-ys[0]
    xpos, ypos = np.meshgrid(xs[:-1]+dx/2, ys[:-1]+dy/2, indexing='ij')
    xpos = xpos.ravel(); ypos = ypos.ravel(); zpos = 0
    wx = dx; wy = dy
    wz = zs.ravel()
    
    if ax==None:
        ax = plt.subplot(111, projection='3d')
    ax.bar3d(xpos,ypos,zpos,wx,wy,wz) #, zsort='averate')
    plt.show()
    
    return zs, xs, ys
```

```python
import matplotlib.pyplot as plt
import numpy as np
from hist3d_my import *
from bar3d_my import *

dx = 0.5; dy = 1
x = np.arange(-2.5, 2.51, dx)
y = np.arange(-1, 1.1, dy)
X, Y = np.meshgrid(x,y)
Z = np.exp(-X**2 - Y**2)

plt.figure(figsize=(10,10))
ax1 = plt.subplot(1,2,1, projection='3d')
bar3d_my(X,Y,Z, wx=dx/2, wy=dy/2, ax=ax1)

plt.figure(figsize=(10,10))
ax2 = plt.subplot(1,2,2, projection='3d')
x, y = np.random.rand(2, 100)
hist3d_my(x,y, Nb=8, ax=ax2)
```
