# 8.13 등고선(Contour)과 화살(Quiver)형 그래프

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

delta = 0.25 #Grid interval
x = np.arange(-2.5,2.5,delta)
y = np.arange(-2.5,2.5,delta)
X,Y = np.meshgrid(x,y) #Grid행렬
Z1 = np.exp(-(X-1)**2 - (Y-1)**2)
Z2 = np.exp(-(X+1)**2 - (Y+1)**2)
Z = Z1-Z2

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
levels = np.arange(-1,1,0.2) #그리고자  하는 등고선들에 대한 z값으로 구성된 1D array
CS = plt.contour(X,Y,Z, levels)
plt.clabel(CS, fontsize=10) #각 등고선에 대한 level값을 삽입
plt.grid()
plt.title('Simplest default contour with labels')

plt.subplot(2,2,2)
(U,V) = (-2*(X-1)*Z1 + 2*(X+1)*Z2, -2*(Y-1)*Z1 + 2*(Y+1)*Z2) #gradient vector
plt.quiver(X,Y, U,V) #vector 화살표를 나타낼 위치 X,Y와 gradient vector값 U,V
plt.grid()
plt.show()

```
