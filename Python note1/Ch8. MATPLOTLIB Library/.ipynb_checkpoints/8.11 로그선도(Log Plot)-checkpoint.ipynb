# 8.11 로그선도(Log Plot)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(tight_layout=True, figsize=(10,5))

plt.subplot(2,2,1)
x = np.logspace(-1,1,1001) #[10**-1, 10**1]구간내 log space의 균등한 1001개 점
y = abs(1j*x/(1+1j*x)) #x에 대한 어떤 복소함수의 절대값
plt.plot(x,y)
plt.grid(True, which="both")

plt.subplot(2,2,2)
plt.semilogx(x,y) #x축 log
plt.grid(True, which="both")

plt.subplot(2,2,3)
plt.semilogy(x,y) #y축 log
plt.grid(True, which="both")

plt.subplot(2,2,4)
plt.loglog(x,y) #x, y축 log
plt.grid(True, which="both")
plt.show()
```
