# 8.1 그래프창을 여러 조각으로 나누는 방법

```python
import matplotlib.pyplot as plt
x = [-0.1, 0.1, 0.2, 0.4, 0.5, 0.7, 1]
y = [2, 1, 0, 0.5, -1, 0, 3]
#주어진 data

fig = plt.figure(tight_layout=True, figsize=(6.4,4.8))
#plt.figure() 함수를 사용하여 그래프창 생성
#tight_layout = True: 앞으로 창을 나누어 각 조각에 그릴 그래프들이 서로 겹치지 않게 설정
#figsize = (6.4,4.8): 그래프창 크기를 6.4x4.8 [inch]로 지정

ax1 = plt.subplot2grid((3,4),(0,0), colspan=4)
#그래프창을 세로 3조각 가로 4조각으로 나눠서 그중 위에서 0번째 위에서 0번째 그래프를 생성
#colspan = 4: (0,0)부터 가로로 4조각을 차지
ax1.plot(x,y,'r.')
#ax1을 'r.' red 점형으로 만든다.

ax2 = plt.subplot2grid((3,4),(1,0), rowspan=2, colspan=2)
#(1,0)부터 시작하여 가로 세로 2조각씩을 차지하는 그래프 생성
ax2.plot(x,y,'b:')
#ax2를 'b:' blue 점선형으로 만든다.

ax3 = plt.subplot2grid((3,4),(1,2), colspan=2)
#(1,2)부터 시작하여 가로로 2조각을 차지하는 그래프를 생성
ax3.plot(x,y,'k')
#ax3를 'k' black 실선형으로 만든다.

ax4 = plt.subplot2grid((3,4),(2,2))
#(2,2)자리에 그래프를 생성
ax4.plot(x,y,'m-x')
#ax4를 'm-x' 분홍색(magenta) 실선(+x)형으로 만든다.

ax5 = plt.subplot2grid((3,4),(2,3))
#(2,3)자리에 그래프를 생성
ax5.plot(x,y,'go-.')
#ax5를 'go-.' 초록색(green) 점(o)실선점(-.)형으로 만든다.
ax5.axis('equal')
#x축과 y축의 scale을 똑같이 설정한다.

plt.show()
#지금까지 만들어진 그래프들을 화면에 출력한다.
```

```python
import matplotlib.pyplot as plt
x = [-0.1, 0.1, 0.2, 0.4, 0.5, 0.7, 1]
y = [2, 1, 0, 0.5, -1, 0, 3]

plt.figure(tight_layout=True, figsize=(6.4,4.8))

plt.subplot2grid((3,4),(0,0), colspan=4)
plt.plot(x,y,'r.')

plt.subplot2grid((3,4),(1,0), rowspan=2, colspan=2)
plt.plot(x,y,'b:')

plt.subplot2grid((3,4),(1,2), colspan=2)
plt.plot(x,y,'k')

plt.subplot2grid((3,4),(2,2))
plt.plot(x,y,'m-x')

plt.subplot2grid((3,4),(2,3))
plt.plot(x,y,'go-.')
plt.axis('equal')
plt.xlim([-1,1.5]), plt.ylim([-1,3])

plt.show()

# 꼭 축을 생성할 필요가 없다면 축을 할당하는 과정을 생략해 coding을 간단히 할 수 있다.
```

```python
import matplotlib.pyplot as plt
x = [-0.1, 0.1, 0.2, 0.4, 0.5, 0.7, 1]
y = [2, 1, 0, 0.5, -1, 0, 3]

plt.figure(tight_layout=True, figsize=(6.4,4.8))

plt.subplot(2,3,1) #그래프창을 행렬 2x3로 나눈 6조각 중 1번째에
plt.plot(x,y,'r.') #적색 . symbol로 그린다

plt.subplot(2,3,2) #2번째에
plt.plot(x,y,'b:') #청색 : line으로 그린다

plt.subplot(2,3,3) #3번째에
plt.plot(x,y,'k') #흑색 선으로 그린다

plt.subplot(2,3,4) #4번째에
plt.plot(x,y,'m-x') #분홍색 - line, x symbol로 그린다

plt.subplot(2,3,5) #5번째에
plt.plot(x,y,'go-.') #녹색 - line, o symbol로 그린다.

plt.subplot(2,3,6) #6번째에
plt.plot(x,y,'y^') #황색 ^symbol로 그린다.

plt.show()

#조각의 크기를 동일하게 할당하고 축에 대한 ~.axis() 명령문을 실행시킬 필요가 없다면
#(모든 축이 같다면) 위와 같이 coding을 간단히 할 수 있다.
```
