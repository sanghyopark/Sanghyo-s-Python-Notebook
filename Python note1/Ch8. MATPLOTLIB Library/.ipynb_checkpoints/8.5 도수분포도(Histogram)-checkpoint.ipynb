# 8.5 도수분포도(Histogram)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(tight_layout=True, figsize=(10,5))
Nx = 50000
m = 100
sigma = 15
ys = m+sigma*np.random.randn(Nx)

plt.subplot(2,2,1)
Nb = 16 #histogram에 대한 구간의 개수
plt.hist(ys,Nb)

plt.subplot(2,2,2)
plt.hist(ys,Nb,density=True) #density=True 총면적이 1이되게

plt.subplot(2,2,3)
ys1=[60 if y<60 else y for y in [140 if y>140 else y for y in ys]]
#60미만은 60으로 140초과는 140으로
ns, bins, patches = plt.hist(ys1, Nb, histtype='step')
#ns: 샘플갯수, bins: 경계값
print(f'ns={ns}\nbins={bins}')
bw = bins[1]-bins[0] #이웃하는 경계값의 차이, 즉 각 막대의 폭
xs = [x+bw/2 for x in bins[0:-1]] #각 구간의 중심에 대한 x좌표들
plt.plot(xs, ns, 'r:')

plt.subplot(2,2,4)
plt.hist(ys,Nb, cumulative=True)
#cumulative=True: 누적도수분포표
plt.show()
```
