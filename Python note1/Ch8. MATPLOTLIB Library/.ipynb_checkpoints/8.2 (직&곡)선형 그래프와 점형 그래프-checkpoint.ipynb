# 8.2 (직&곡)선형 그래프와 점형 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(tight_layout=True, figsize=(8.8,2.4))
x = [6.1, 5.8, 5.7, 5.6, 5.5, 5.9, 5.3, 5.2] #Unemployment_Rate
y = [1500, 1520, 1525, 1540, 1545, 1508, 1560, 1555] #Stock_Index_Price

plt.subplot(1,2,1)
plt.plot(x,y,'r', x,y,'b.')
plt.xlabel('Unemployment Rate', fontsize=10)
plt.ylabel('Stock Index Price', fontsize=10)
plt.title('Unemployment Rate vs. Stock Index Price', fontsize=10)

plt.subplot(1,2,2)
xa = np.array(x); ya = np.array(y);
sorted_ind = np.argsort(xa) 
#x,y data들을 nparray형으로 변환해서 np.argsort()함수를 이용해 xa를 정렬하는 index를 생성하고
xa = xa[sorted_ind]; ya = ya[sorted_ind]
#이를 이용하여 xa와 ya를 정렬시킨다
plt.plot(xa,ya,'b')
plt.xlabel('Unemployment Rate', fontsize=10)
plt.ylabel('Stock Index Price', fontsize=10)
plt.title('Unemployment Rate vs. Stock Index Price', fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(tight_layout=True, figsize=(8.6,4.8))
f = lambda t: np.exp(-t)*np.cos(2*np.pi*t) #함수 설정
t1 = np.arange(0,2,0.1) #(0,2)구간에 0.1간격의 수열 생성
t2 = np.arange(0,2,0.01) #(0,2)구간에 0.01간격의 수열 생성

plt.subplot(2,2,1) #그래프창을 행렬 2x2로 나눈 조각중 첫번째
plt.plot(t1, f(t1), 'o', markersize=4) #o symbol 크기를 4로 한다.
plt.title('Plot 1', fontsize=10)

plt.subplot(2,2,2) #그래프창을 행렬 2x2로 나눈 조각중 두번째
plt.plot(t2, f(t2), '-')
plt.title('Plot 2', fontsize=10)

plt.subplot(2,1,2) #그래프창을 행렬 2x1로 나눈 조각중 두번째
plt.plot(t1, f(t1), 'o', label='f(t1)') #o symbol로 그리고 label을 f(t1)으로 정한다.
plt.plot(t2, f(t2), '-', label='f(t2)') #- line으로 그리고 label을 f(t2)으로 정한다.
plt.xlabel('time t[s]')
plt.ylabel('f(t)')
plt.title('Plot 3', fontsize=10)
plt.legend()
#위에서 정한 그래프의 label을 legend, 즉 각 그래프를 구분할 수 있도록 설명하는 그래프 기호
#일람표를 띄운다.

plt.show()
```

## (예제 8.1) 두 선형그래프의 사이 또는 한 선형그래프와 x축 사이를 채우기 - plt.fill_between() 함수

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.arange(0,10,0.1)

plt.figure(tight_layout=True, figsize=(8.6,4.8))

plt.subplot(2,3,1)
y = 0.1*x + 0.3
plt.plot(x,y,'r')
plt.fill_between(x, 0.9*y, 1.1*y, alpha=0.2) #alpha: 색깔의 농담
#x에 대한 0.9*y와 1.1*y의 그래프 사이를 채운다
plt.ylim(0,2)

plt.subplot(2,3,2)
y1 = 2 - 0.15*x
plt.fill_between(x, y, y1, where=(y<y1), color='blue', alpha=0.1)
# y<y1인 x축 구간에 대해 y와 y1 그래프들 사이를 청색으로 채운다
plt.fill_between(x, y, y1, where=(y>y1), color='red', alpha=0.1)
# y>y1인 x축 구간에 대해 y와 y1 그래프 사이를 적색으로 채운다.
plt.ylim(0,2)

plt.subplot(2,3,3)
plt.plot(x,y,':', label='y=0.1x+0.3')
plt.plot(x,y1,':', label='y1=2-0.15x')

y2 = np.minimum(y,y1) #y와 y1중 작은값들로 구성된 수열을 y2로 만든다.
plt.fill_between(x, y2, color='green', alpha=0.1)
plt.ylim(0,2)
plt.legend()

plt.show()
```
