# 8.14 Stream Plot

```python
import matplotlib.pyplot as plt
import numpy as np

Y,X = np.mgrid[0:2:0.2, 0:2:0.2]
#np.meshgrid()       
U = 2*X; V = -2*Y
speed = np.sqrt(U**2 + V**2)
strm = plt.streamplot(X,Y, U,V, color=speed, cmap='autumn')
#   .
plt.colorbar(strm.lines)
plt.title('Stream plot with varying color')
plt.show()
```
